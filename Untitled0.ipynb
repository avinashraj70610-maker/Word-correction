```python
dictionary = ["receive","believe","friend","learning","machine","python","science","data",
              "school","college","project","analysis"]
```

```python
print(dictionary)
```

```python
def char_ngrams(word, n=2):
    word = word.lower()
    return [word[i:i+n] for i in range(len(word) - n + 1)]
```

```python
# jaccard Similarity Function
def jaccard_similarity(a,b):
  set1=set(a)
  set2=set(b)
  return len(set1 & set2)/len(set1 | set2)
```

```python
def improved_correction(word,dictionary,n=2):
  word_ngrams=char_ngrams(word,n)
  best_match=None
  best_score = 0

  for correct_word in dictionary:
    correct_ngrams = char_ngrams(correct_word,n)
    score = jaccard_similarity(word_ngrams,correct_ngrams)

    if score>best_score:
      best_score = score
      best_match = correct_word

      return best_match,best_score
```

```python
wrong_words = ["recieve","friend","learning","machin","science"]
for w in wrong_words:
  match,score = improved_correction(w,dictionary)
  print(f"(w)--->{match} (score:{score:2f})")

```
